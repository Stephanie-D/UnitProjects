```python
DNA_Manatee=open("/content/drive/My Drive/Colab Notebooks/molecular genetics/4a_CSN2_DNA.txt").read()
DNA_Dolphins= open("/content/drive/My Drive/Colab Notebooks/molecular genetics/4b_CSN2_DNA.txt").read()
Cow_DNA=open("/content/drive/My Drive/Colab Notebooks/molecular genetics/4c_CSN2_DNA.txt").read()
```

```python
#Dolphins RNA Transcription
RNA_Dolphins=""
print(RNA_Dolphins)

transcribe={'A':'U','T':'A','G':'C','C':'G'}

for i in DNA_Dolphins:
  if i is 'A':
    RNA_Dolphins += transcribe['A']
  elif i is 'T':
    RNA_Dolphins += transcribe['T']
  elif i is 'G':
    RNA_Dolphins += transcribe['G']
  else:
      RNA_Dolphins += transcribe['C']

print(RNA_Dolphins)
```

```python
#Cow RNA Transcribtion
RNA_Cow=""
print(RNA_Cow)

transcribe={'A':'U','T':'A','G':'C','C':'G'}

for i in Cow_DNA:
  if i is 'A':
    RNA_Cow += transcribe['A']
  elif i is 'T':
    RNA_Cow += transcribe['T']
  elif i is 'G':
    RNA_Cow += transcribe['G']
  else:
      RNA_Cow += transcribe['C']

print(RNA_Cow)
```

```python
#Manatee RNA Transcribtion
RNA_Manatee=""
print(RNA_Manatee)

transcribe={'A':'U','T':'A','G':'C','C':'G'}

for i in DNA_Manatee:
  if i is 'A':
    RNA_Manatee += transcribe['A']
  elif i is 'T':
    RNA_Manatee += transcribe['T']
  elif i is 'G':
    RNA_Manatee += transcribe['G']
  else:
      RNA_Manatee += transcribe['C']

print(RNA_Manatee)
```

```python
#Dolphins Protien Translation
codon=""
Dolphins_protein= ""
counter =0

for i in RNA_Dolphins:
  if len(codon) == 3:
    Dolphins_protein += translation[codon] 
    codon ="" 
    codon += i
  else:
    codon += i

print(Dolphins_protein)
```

```python
#Cow Protien Translation
codon=""
Cow_protein= ""
counter =0

for i in RNA_Cow:
  if len(codon) == 3:
    Cow_protein += translation[codon] 
    codon ="" 
    codon += i
  else:
    codon += i

print(Cow_protein)
```

```python
#Manatee Protien Translation
codon=""
Manatee_protein= ""
counter =0

for i in RNA_Manatee:
  if len(codon) == 3:
    Manatee_protein += translation[codon] 
    codon ="" 
    codon += i
  else:
    codon += i

print(Manatee_protein)
```

```python
#Dolphins and Cow Simalarity percentage(protein)
difference_a=0
similarites_a =0
for counter,value in enumerate(Dolphins_protein):
   if value == Cow_protein[counter]:
      similarites_a+=1
   else:
      difference_a+=1
percent_sim_a = similarites_a/len(Dolphins_protein)*100
percent_diff_a = difference_a/len(Dolphins_protein)*100
print(percent_sim_a)
print(percent_diff_a)
```
